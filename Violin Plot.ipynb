```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_excel("ESD.xlsx")
df = pd.DataFrame(data)
#print(df)
plt.violinplot(df["Age"],showmedians = True)
plt.title('Violin Plot of Employee Ages')
plt.ylabel('Age')
plt.show()
```
