```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_excel("ESD.xlsx")
df = pd.DataFrame(data)
#print(df)
plt.hist(df["Age"],bins = 15, edgecolor = "black")
plt.show()

# This Python script reads data from an Excel file and creates a histogram to visualize the distribution of ages.
# It uses Pandas for data manipulation and Matplotlib for plotting.
# The histogram includes 15 bins with black edges for better clarity.
```
