```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_excel("expense.xlsx")
df = pd.DataFrame(data)
grouped_by = df.groupby("Payment Mode")["Amount"].sum()
print(grouped_by)
ex = [0,0.1,0]
plt.pie(grouped_by.values, labels = grouped_by.index, autopct = "%.2f", explode = ex, startangle = 180)
plt.savefig("piee3e.png", facecolor = "green",pad_inches = 0.2)
plt.show()
#print(df)
```
