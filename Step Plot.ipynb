```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_excel("expense.xlsx")
df = pd.DataFrame(data)
#print(df)
group = df.groupby("Category").agg({"Amount":"sum"})
print(group)
plt.step(group.index, group["Amount"], where = "mid", marker = "o")
plt.title('Expense with Respect to Amount')
plt.xlabel('Category')
plt.ylabel('Amount')
plt.show()
```
