```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data = pd.read_excel("ESD.xlsx")
df = pd.DataFrame(data)


#To Count the frequency of employees by age
# Scatter plot: Age vs. Frequency
plt.figure(figsize=(12, 8)) 
plt.scatter(df['Age'], range(len(df)), alpha=0.6, color='blue', edgecolor='k')

plt.title('Scatter Plot of Employee Ages')
plt.xlabel('Age')
plt.ylabel('Employee Count')
plt.grid(True)
plt.show()


# To Count the number of employees by age
age_counts = df['Age'].value_counts().sort_index()

# Scatter plot: Age vs. Count
plt.figure(figsize=(8, 6))
plt.scatter(age_counts.index, age_counts.values, color='green', edgecolor='k')

# Plot customization
plt.title('Employee Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.grid(True)
plt.show()
```
