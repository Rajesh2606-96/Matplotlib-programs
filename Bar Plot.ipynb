```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel("expense.xlsx")
#print(data.isnull().sum()) # to check the null values
#print(data.info()) # to check the data type

df = pd.DataFrame(data)
grouped_by = df.groupby("Payment Mode")["Amount"].sum()
#print(grouped_by)
plt.bar(grouped_by.index,grouped_by.values,color = "lightgray",edgecolor = "black")
plt.xlabel("Payment Mode",fontsize = 15)
plt.ylabel("Amount",fontsize = 15)
plt.title("Expenses by Payment Mode",fontsize = 15)
plt.savefig("piee.png", facecolor = "black")
plt.show()

# This code reads data from an Excel file (ESD.xlsx) and creates a histogram of the "Age" column using Matplotlib. 
# The histogram is displayed with 15 bins and black edges for better visualization.
# The code demonstrates basic data analysis and visualization with Pandas and Matplotlib.
```
